```python
#All in one step
import pandas as pd

def process_exons_split_attributes_and_filter(input_gff3_file, output_csv_file):
    """
    Load a GFF3 file, filter for exons, split the 'attributes' column into separate columns,
    filter for 'protein_coding' gene_type, and save the result to a CSV file.

    Parameters:
    input_gff3_file (str): The file path for the input GFF3 file, including file name.
    output_csv_file (str): The file path for the output CSV file, including file name.
    """
    # Loading the GFF3 file with specified headers
    gff3_df = pd.read_csv(input_gff3_file, sep='\t', comment='#', header=None,
                          names=['seqid', 'source', 'type', 'start', 'end',
                                 'score', 'strand', 'phase', 'attributes'])

    # Filtering for exons
    exon_df = gff3_df[gff3_df['type'] == 'exon']

    # Spliting of the 'attributes' column into separate columns
    attributes_df = exon_df['attributes'].str.split(';', expand=True)
    attributes_expanded = pd.DataFrame()

    for col in attributes_df.columns:
        # Further split each attribute into key-value pairs
        split_columns = attributes_df[col].str.split('=', expand=True)
        if len(split_columns.columns) == 2:
            # Use the first part as the key and the second part as the value
            key = split_columns.iloc[:, 0].str.strip()
            value = split_columns.iloc[:, 1]
            attributes_expanded[key.iat[0]] = value

    # To Merge the expanded attributes back into the exon DataFrame
    data_final = pd.concat([exon_df.drop(columns=['attributes']), attributes_expanded], axis=1)

    # To Filter for 'protein_coding' gene_type
    protein_coding_df = data_final[data_final['gene_type'] == 'protein_coding']

    # Saving the processed DataFrame to the specified CSV file
    protein_coding_df.to_csv(output_csv_file, index=False)
    print(f"Processed exon data saved to {output_csv_file}")


input_gff3_file = 'gencode.v45.annotation.gff3'
output_csv_file = 'new_processed_exons_protein_coding.csv'
process_exons_split_attributes_and_filter(input_gff3_file, output_csv_file)
```

```python
import pandas as pd
import os

def find_first_overlapping_exon(rbp_start, rbp_end, exons_starts, exons_ends):
    low, high = 0, len(exons_starts) - 1
    while low <= high:
        mid = (low + high) // 2
        exon_start, exon_end = exons_starts[mid], exons_ends[mid]

        if exon_end < rbp_start:
            low = mid + 1
        elif exon_start > rbp_end:
            high = mid - 1
        else:
            while mid > 0 and exons_ends[mid - 1] >= rbp_start:
                mid -= 1
            return mid
    return None

def process_overlap_findings(in_dir, out_dir, exon_annotation_csv):
    # To Ensure the output directory exists
    os.makedirs(out_dir, exist_ok=True)
    
    # Loading exon annotations
    exons_df = pd.read_csv(exon_annotation_csv)
    exons_df.rename(columns={'seqid': 'chromosome'}, inplace=True)
    column_names = ['chromosome', 'chromStart', 'chromEnd', 'signalValue', 'ID', 'Parent', 'gene_id', 'transcript_id', 'gene_type',
                    'gene_name', 'transcript_type', 'transcript_name', 'exon_number', 'exon_id']

    files = os.listdir(in_dir)
    for file in files:
        if not file.endswith('.bed'):
            continue  # Skip non-BED files

        infile = os.path.join(in_dir, file)
        rbp_sites_df = pd.read_csv(infile, sep='\t', header=None)
        rbp_sites_df.rename(columns={0: 'chr', 1: 'chromStart', 2: 'chromEnd', 3: 'name', 4: 'score', 5: 'strand', 6: 'signalValue', 7: 'pValue', 8: 'qValue', 9: 'peak'}, inplace=True)
        rbp_sites_df.sort_values(['chr', 'chromStart'], inplace=True)
        outfile = os.path.join(out_dir, file.replace(".bed", "_out.csv"))
        rbp_sites_df.rename(columns={'chr': 'chromosome'}, inplace=True)

        overlapping_rbp_sites = []

        for chromosome in rbp_sites_df['chromosome'].unique():
            if chromosome not in exons_df['chromosome'].unique():
                continue

            rbp_sites_chr = rbp_sites_df[rbp_sites_df['chromosome'] == chromosome]
            exons_chr = exons_df[exons_df['chromosome'] == chromosome]

            rbp_starts = rbp_sites_chr['chromStart'].to_numpy()
            rbp_ends = rbp_sites_chr['chromEnd'].to_numpy()
            rbp_signals = rbp_sites_chr['signalValue'].to_numpy()
            exons_starts = exons_chr['start'].to_numpy()
            exons_ends = exons_chr['end'].to_numpy()

            for i, (rbp_start, rbp_end, rbp_signal) in enumerate(zip(rbp_starts, rbp_ends, rbp_signals)):
                first_overlap_index = find_first_overlapping_exon(rbp_start, rbp_end, exons_starts, exons_ends)
                if first_overlap_index is not None:
                    for j in range(first_overlap_index, len(exons_starts)):
                        exon_start, exon_end = exons_starts[j], exons_ends[j]
                        if rbp_start <= exon_end and rbp_end >= exon_start:
                            overlapping_rbp_sites.append({
                                'chromosome': chromosome,
                                'chromStart': rbp_start,
                                'chromEnd': rbp_end,
                                'signalValue': rbp_signal,
                                **exons_chr.iloc[j][['ID', 'Parent', 'gene_id', 'transcript_id', 'gene_type', 'gene_name', 'transcript_type', 'transcript_name', 'exon_number', 'exon_id']].to_dict()
                            })
                            break
        
        
        overlapping_rbp_sites_df = pd.DataFrame(overlapping_rbp_sites)
        overlapping_rbp_sites_df.to_csv(outfile, index=False)

        print(f"\nTotal RBP sites situated inside exons: {len(overlapping_rbp_sites)}")
        print(f"{infile} is done!")
        print("****************************************")

# Example usage
in_dir = 'input-rbp'
out_dir = 'overlap_RBPs_April'
exon_annotation_csv = 'new_last_gencode_exons_sorted.csv'
process_overlap_findings(in_dir, out_dir, exon_annotation_csv)
```

```python
#Of the 145 RNA-binding protein (RBP) experiments, six involved repeated studies. We analyzed each RBP's results individually and 
# then compiled a final dataset from only those findings that were consistent across repeated experiments (overlapped), ensuring the reliability of our analysis

import pandas as pd

def merge_replicate_overlaps(replicate1_path, replicate2_path, output_path):
    """
    Finds overlap regions between two files of biological replicates and saves the overlaps to a new CSV file.

    Parameters:
    replicate1_path (str): Path to the first CSV file of replicate data.
    replicate2_path (str): Path to the second CSV file of replicate data.
    output_path (str): Path to save the output CSV file containing the overlapping entries.
    """
    # Load the CSV files for the biological replicates
    replicate1_df = pd.read_csv(replicate1_path)
    replicate2_df = pd.read_csv(replicate2_path)

    # Ensuring the data is sorted can sometimes improve merge efficiency
    replicate1_df.sort_values(by=['chromosome', 'chromStart', 'chromEnd'], inplace=True)
    replicate2_df.sort_values(by=['chromosome', 'chromStart', 'chromEnd'], inplace=True)

    # To Merge on 'chromosome' to ensure comparisons are made within the same chromosome
    merged_df = pd.merge(replicate1_df, replicate2_df, on='chromosome', suffixes=('_r1', '_r2'))

    # Vectorized check for overlaps
    overlaps = merged_df[
        (merged_df['chromStart_r1'] <= merged_df['chromEnd_r2']) &
        (merged_df['chromEnd_r1'] >= merged_df['chromStart_r2'])
    ]

    # Extracting relevant information or cleaning up the DataFrame as needed

    # Saving the overlaps to a new CSV file
    overlaps.to_csv(output_path, index=False)
    print(f"Overlapping entries have been saved to '{output_path}'.")

replicate1_path = 'ENCFF037TVC_TARDBP2_out.csv'
replicate2_path = 'ENCFF593RED_TARDBP_out.csv'
output_path = 'overlapping_entries_TARDBP.csv'
merge_replicate_overlaps(replicate1_path, replicate2_path, output_path)
```

```python
# I used the mean of SignalValue of replicate1 and 2
import pandas as pd
import os

def process_files_and_update_signal_values(input_folder_path, output_folder_path):
    """
    Processes CSV files in a specified folder by calculating the mean of signalValue_r1 and signalValue_r2,
    keeps only the first 14 columns, renames columns by removing '_r1' suffix, and saves the processed files
    to a new output folder.

    Parameters:
    - input_folder_path: Path to the folder containing the input CSV files.
    - output_folder_path: Path to the folder where processed CSV files will be saved.
    """
    # Ensuring the output folder exists
    os.makedirs(output_folder_path, exist_ok=True)

    # Iterate through each file in the input folder
    for filename in os.listdir(input_folder_path):
        if filename.endswith(".csv"):
            input_file_path = os.path.join(input_folder_path, filename)
            output_file_path = os.path.join(output_folder_path, filename)
            print(f"Processing {input_file_path}...")

            # Read the CSV file
            df = pd.read_csv(input_file_path)

            # Calculating the mean of signalValue_r1 and signalValue_r2, update signalValue_r1
            df['signalValue_r1'] = df[['signalValue_r1', 'signalValue_r2']].mean(axis=1)

            # Keeping only the first 14 columns
            df = df.iloc[:, :14]

            # Removing "_r1" from the column headers
            df.columns = [col.replace("_r1", "") for col in df.columns]

            # Saving the processed DataFrame to a new file in the output folder
            df.to_csv(output_file_path, index=False)
            
            print(f"Updated data saved to {output_file_path}")

input_folder_path = 'beeeiiin/'  
output_folder_path = 'beeeiiin/'  
process_files_and_update_signal_values(input_folder_path, output_folder_path)

print("All files have been processed.")
```

```python
#This function, sum_signal_values_by_gene_id, correctly process each CSV file by summing up signalValue for each unique gene_id.
#It creates an output file for each input file, where each output file contains two columns: gene_id and the corresponding sum of signalValue. 

import pandas as pd
import os

def sum_signal_values_by_gene_id(input_dir, output_dir):
    """
    For each CSV file in the input directory, this function sums the 'signalValue' for each unique 'gene_id'.
    It then writes these sums along with the 'gene_id' to a new CSV file in the output directory.
    If an input file is empty, an output file with just the headers 'gene_id' and 'signalValue' is created.

    Parameters:
    input_dir (str): The directory containing the input CSV files.
    output_dir (str): The directory where the output CSV files will be saved.
    """
    # To Ensure the output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # List all CSV files in the input directory
    rbp_files = [f for f in os.listdir(input_dir) if f.endswith('.csv')]

    for rbp_file in rbp_files:
        input_file_path = os.path.join(input_dir, rbp_file)
        output_file_path = os.path.join(output_dir, f"summed_{rbp_file}")
        
        # Attempt to load the current RBP file
        df = pd.read_csv(input_file_path)

        # If the DataFrame is empty, create an output file with only headers
        if df.empty:
            empty_df = pd.DataFrame(columns=['gene_id', 'signalValue'])
            empty_df.to_csv(output_file_path, index=False)
            print(f"Found empty file: {rbp_file}. Header-only file created: {output_file_path}")
            continue

        # Sum 'signalValue's for each unique 'gene_id'
        summed_df = df.groupby('gene_id')['signalValue'].sum().reset_index()

        # Saving the summed data to a new CSV file
        summed_df.to_csv(output_file_path, index=False)
        print(f"Processed and saved: {output_file_path}")


input_dir = 'Birthday/'  
output_dir = 'Birthdayparty1'  
sum_signal_values_by_gene_id(input_dir, output_dir)
```

```python
#To achieve the code that converts the CSV annotation file into the GFF3 format within a function, 
#we defined a function named csv_to_gff3. This function takes the path to the input CSV file and the desired output GFF3 file path as parameters.
#This conversion is essential for creating a local GFF file filtered by exons of protein-coding genes,
#allowing us to align Ribo-seq data precisely against these gene annotations. This ensures accurate measurement of translation activity, crucial for our analysis.
```

```python
import pandas as pd
import gzip
import shutil
import os

def csv_to_gff3(input_csv_path, output_gff3_path):
    """
    Converts a CSV file containing gene annotations into a GFF3 format file.
    
    Parameters:
    input_csv_path (str): The file path to the input CSV file with annotations.
    output_gff3_path (str): The file path where the output GFF3 file should be saved.
    """
    # Load the CSV file
    data = pd.read_csv(input_csv_path)

    # Construct the 'attributes' column with all necessary details
    data['attributes'] = data.apply(
        lambda x: f"ID={x['ID']};Parent={x['Parent']};gene_id={x['gene_id']};"
                  f"transcript_id={x['transcript_id']};gene_type={x['gene_type']};"
                  f"gene_name={x['gene_name']};transcript_type={x['transcript_type']};"
                  f"transcript_name={x['transcript_name']};exon_number={x['exon_number']};"
                  f"exon_id={x['exon_id']};level={x['level']};"
                  f"transcript_support_level={x.get('transcript_support_level', '')};"
                  f"tag={x.get('tag', '')};havana_transcript={x.get('havana_transcript', '')}",
        axis=1
    ).str.replace(';;', ';')  # Clean up potential trailing semicolons

    # Format the DataFrame for GFF3
    gff3_columns = ['seqid', 'source', 'type', 'start', 'end', 'score', 'strand', 'phase', 'attributes']
    gff3_data = data[gff3_columns].fillna({
        'source': '.',  # Fill in with actual data or a placeholder
        'score': '.',
        'strand': '+',  # Adjust based on your data
        'phase': '.'
    })

    # Convert 'start' and 'end' to integers
    gff3_data['start'] = gff3_data['start'].astype(int)
    gff3_data['end'] = gff3_data['end'].astype(int)

    # Save to GFF3 file
    gff3_data.to_csv(output_gff3_path, sep='\t', index=False, header=False, quoting=3)

    print(f"GFF3 file saved to {output_gff3_path}")
    gz_file_path = output_gff3_path + '.gz'
    with open(output_gff3_path, 'rb') as f_in:
        with gzip.open(gz_file_path, 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)
    print(f"GFF3 file compressed and saved to {gz_file_path}")
    
# Example usage
input_csv_path = 'new_last_gencode_exons_sorted.csv'  # Update this to your actual file path
output_gff3_path = 'test-proteinfiltered_output.gff3'  # Update this to your desired output path
csv_to_gff3(input_csv_path, output_gff3_path)
```

```python
#TPM Normalisation
import pandas as pd

# Loading the merged CSV file into a DataFrame
df = pd.read_csv('merged_output_file.csv')


# Calculating reads per kilobase (RPK)
df['RPK'] = df['Mean_Counts'] / (df['Length'] / 1000)

# Calculating per million scaling factor
scaling_factor = df['RPK'].sum() / 1e6

# Calculate TPM
df['TPM'] = df['RPK'] / scaling_factor


# Defining the path for the DataFrame with TPM values
output_tpm_path = 'Normalized_shScramble_output_file_with_TPM4.csv'

# To Save the DataFrame to a new CSV file
df.to_csv(output_tpm_path, index=False)

print(f"Data with TPM normalization saved to {output_tpm_path}")
```

```python
#Upon processing the data for all four conditions (shScramble, shDDX41, CX5461, and DMSO), we constructed the final dataset. 
#The dataset structure includes the first column populated with 20,049 gene IDs. Columns two through 140 contain RBP signal data corresponding to each gene ID. 
#The last four columns represent the FeatureCounts of these genes, which have been normalized using TPM,
#ensuring accurate representation of gene expression levels across different samples.
```

```python
#First step to add RBPs data merged
import pandas as pd
import numpy as np
import os

# Load the main file containing Geneid
featurecounts_path = 'chromosomes_only/Featurecounts_proteinfiltered_length_output_file.csv'
featurecounts_df = pd.read_csv(featurecounts_path)

# Initialize a dictionary to hold data for the final DataFrame
final_data = {'Geneid': featurecounts_df['Geneid'].values}
columns_dtype = {}

rbp_folder_path = 'chr-summed_output_rbp'
rbp_files = [f for f in os.listdir(rbp_folder_path) if f.endswith('.csv')]

# Pre-fill the dictionary with Geneids as keys and a list full of zeros as values
for rbp_file in rbp_files:
    column_name = rbp_file.replace('.csv', '')
    final_data[column_name] = np.zeros(len(featurecounts_df), dtype=float)
    columns_dtype[column_name] = 'float64'

# Process each RBP file
for rbp_file in rbp_files:
    rbp_path = os.path.join(rbp_folder_path, rbp_file)
    rbp_df = pd.read_csv(rbp_path)
    column_name = rbp_file.replace('.csv', '')
    
    # Map 'signalValue's to 'Geneid'
    for i, gene_id in enumerate(final_data['Geneid']):
        value = rbp_df.loc[rbp_df['gene_id'] == gene_id, 'signalValue'].sum()  # Sum in case of multiple entries
        if not pd.isnull(value):
            final_data[column_name][i] = value

# Create the final DataFrame
final_df = pd.DataFrame(final_data)

# Specify dtype explicitly to avoid FutureWarning
for col, dtype in columns_dtype.items():
    final_df[col] = final_df[col].astype(dtype)

# Save the final DataFrame to a new CSV file
final_output_path = 'best_finallllllll_output2.csv'
final_df.to_csv(final_output_path, index=False)
```

```python
#Then to add last 4 columns (translation outcome for each condition)
import pandas as pd

# Load the final output file
final_df = pd.read_csv('best_finallllllll_output_April.csv')

conditions = ['shScramble', 'shDDX41', 'DMSO', 'CX5461']
condition_files = [
    'chromosomes_only/shScramble/Normalized_shScramble_output_file_with_TPM4.csv',
    'chromosomes_only/shDDX41/Normalized4_shDDX41_output_file_with_TPM.csv',
    'chromosomes_only/DMSO/Normalized4_DMSO_output_file_with_TPM.csv',
    'chromosomes_only/CX5461/Normalized_CX5461_output_file_with_TPM.csv',
]

for condition, file_path in zip(conditions, condition_files):
    # Load the condition file
    condition_df = pd.read_csv(file_path)
    
    # Extract Geneid and TPM columns (assuming the column is named 'TPM'; adjust if necessary)
    tpm_data = condition_df[['Geneid', 'TPM']].rename(columns={'TPM': f'TPM_{condition}'})
    
    # Merge the TPM data into the final DataFrame
    final_df = pd.merge(final_df, tpm_data, on='Geneid', how='left')

# Replace NaN values with 0 (assuming you want to treat missing TPM data as 0)
final_df.fillna(0, inplace=True)


final_df.to_csv('ready_file_for_ML_April.csv', index=False)
```
